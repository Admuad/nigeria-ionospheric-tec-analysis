# 06 - Report Graphs

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
ROOT=Path.cwd().parent
df=pd.read_csv(ROOT/'data'/'processed'/'master_dataset.csv')
df['date']=pd.to_datetime(df['date'])
```

```python
annual=df.groupby(df['date'].dt.year)['daily_mean_tec'].mean()
annual.plot(kind='bar')
plt.ylabel('Mean TEC')
plt.show()
```

```python
monthly=df.groupby(df['date'].dt.month)['daily_mean_tec'].mean()
monthly.plot(marker='o')
plt.grid();plt.show()
```

```python
plt.hist(df['daily_mean_tec'],bins=30)
plt.xlabel('TEC');plt.show()
```
