# 05 - Final Figures

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ROOT=Path.cwd().parent
df=pd.read_csv(ROOT/'data'/'processed'/'master_dataset.csv')
df['date']=pd.to_datetime(df['date'])
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['date'],df['daily_mean_tec'])
plt.title('Daily Mean TEC (2022-2024)')
plt.grid();plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[['daily_mean_tec','ssn','kp_mean','ap_daily','f107_obs','dst_daily_mean']].corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

```python
df['month']=df['date'].dt.month
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='month',y='daily_mean_tec')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['f107_obs'],df['daily_mean_tec'],alpha=.6)
plt.xlabel('F10.7');plt.ylabel('TEC');plt.grid();plt.show()
```
